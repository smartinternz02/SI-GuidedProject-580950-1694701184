```python
import pandas as pd
csv_load = 'House Price India.csv'
df = pd.read_csv(csv_load)
df.head()

!pip install seaborn matplotlib
import seaborn as sns
import matplotlib.pyplot as plt

# Histogram of Price
plt.figure(figsize=(10, 6))
sns.histplot(df['Price'], bins=30, kde=True)
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Histogram of Price')
plt.show()

# Scatter Plot between Number of Bedrooms and Price
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='number of bedrooms', y='Price')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Price')
plt.title('Scatter Plot: Number of Bedrooms vs. Price')
plt.show()

# Correlation Matrix
correlation_matrix = df.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5)
plt.title('Correlation Matrix')
plt.show()

descriptive_stats = df.describe()
descriptive_stats

missing_values = df.isnull().sum()
# Interpolation (e.g., for time series data)
df['Price'].interpolate(method='linear', inplace=True)
missing_values_after_handling = df.isnull().sum()
print(df.head())
```
